```python
path = "D:/DS Projects/My_Crawl_Data/Data/"
df0 = pd.read_csv(path + "zingmp3_0-2k.csv")
df0 = df0.iloc[:2001,2:]
df0 = df0[["title","singer","Genre","Lyric"]]
df0_y = df0.dropna()

df0_na = df0.isnull()
df0_na = df0[(df0_na.Genre == True) | (df0_na.Lyric == True)] [["title","singer" ]]
df0_y
```
